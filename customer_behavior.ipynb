```python
df['Review Rating'] = df.groupby('Category')["Review Rating"].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.rename(columns={"purchase_amount_(usd)":"purchase_amount"}, inplace=True)
```

```python
df.columns
```

```python
#create a new column
age_lables = ["Young Adult", "Adult", "Middle Aged", "Senior"]
df['age_group'] = pd.qcut(df['age'], q= 4, labels=age_lables)
```

```python
df[["age_group","age"]].head(10)
```

```python
#getting all unique values from Frequency of purchases so that I can convert them from text to numbers 
df["frequency_of_purchases"].unique()
```

```python
frequency_of_purchases_labels = {'Fortnightly':14, 'Weekly':7, 'Annually':365, 'Quarterly':90, 'Bi-Weekly':14,
       'Monthly':30, 'Every 3 Months':90}
df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_of_purchases_labels)
```
